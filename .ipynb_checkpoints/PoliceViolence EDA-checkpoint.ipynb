```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
policekillings = pd.read_csv('data/police_killings.csv')
```

```python
policekillings.columns = policekillings.columns.str.replace("'",'').str.replace(' ','_')
policekillings.columns = policekillings.columns.str.replace('Street_Address_of_Incident','Address').str.replace('Zipcode','Zip')
policekillings.columns = policekillings.columns.str.replace('_(Source:_WaPo)','')
policekillings = policekillings[policekillings.columns.drop(list(policekillings.filter(regex='Unnamed')))]
```

```python
policekillings = policekillings.rename(columns={'Geography_(via_Trulia_methodology_based_on_zipcode\
_population_density:_http://jedkolko.com/wp-content/uploads/2015/05/full-ZCTA-urban-suburban-rural-\
classification.xlsx_)': "Geography"})
```

```python
policekillings.columns
```

```python
policekillings.dropna(axis = "rows",how="all",inplace=True)
```

```python
policekillings
```

```python
killingsaddress = policekillings.loc[:,['Victims_name','Address', 'City', 'State', 'Zip']]
killingsaddress = killingsaddress.dropna(axis=0, how='any')
killingsaddress.Zip = killingsaddress.Zip.astype('int64')
killingsaddress.to_csv('KA.csv')
```

```python
ka_geocode = pd.read_csv('KA_geocode.csv')
ka_geocode
```

```python
PK_geocode = pd.merge(policekillings, ka_geocode, on='Victims_name')
```

```python
PK_geocode['Symptoms_of_mental_illness?'].replace({"Unkown": "Unknown", "unknown": "Unknown"}, inplace=True)
PK_geocode['Symptoms_of_mental_illness?'].value_counts()
```

```python
PK_geocode['Victims_race'].replace({"Unknown race": "Unknown Race"}, inplace=True)
PK_geocode['Victims_race'].value_counts()
```

```python
PK_geocode = PK_geocode[PK_geocode.columns.drop(list(PK_geocode.filter(regex='_y')))]
PK_geocode.columns = PK_geocode.columns.str.replace("_x",'')
PK_geocode.columns = [x.replace("_(Source:_WaPo)","") for x in PK_geocode.columns]
PK_geocode.columns = [x.replace("Date_of_Incident_(month/day/year)","Date") for x in PK_geocode.columns]
```

```python
PK_geocode.shape
```

```python
PK_geocode.dropna(subset = ['Symptoms_of_mental_illness?'],inplace=True)
PK_geocode.dropna(subset = ['Body_Camera'],inplace=True)
```

```python
PK_geocode['Body_Camera'] = PK_geocode['Body_Camera'].str.replace('no','No')\
.str.replace('Bystander Video','No').str.replace('Surveillance Video','No')
```

```python
PK_geocode['Symptoms_of_mental_illness?'].value_counts()
```

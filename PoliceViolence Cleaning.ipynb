```python
ka_geocode = pd.read_csv('KA_geocode.csv')
ka_geocode
```

```python
PK_geocode = pd.merge(policekillings, ka_geocode, on='Victims_name')
```

```python
PK_geocode['Symptoms_of_mental_illness?'].replace({"Unkown": "Unknown", "unknown": "Unknown","Unknown ": "Unknown"}, inplace=True)
PK_geocode['Symptoms_of_mental_illness?'].value_counts()
```

```python
PK_geocode['Victims_race'].replace({"Unknown race": "Unknown Race"}, inplace=True)
PK_geocode['Victims_race'].value_counts()
```

```python
PK_geocode = PK_geocode[PK_geocode.columns.drop(list(PK_geocode.filter(regex='_y')))]
PK_geocode.columns = PK_geocode.columns.str.replace("_x",'')
PK_geocode.columns = [x.replace("_(Source:_WaPo)","") for x in PK_geocode.columns]
PK_geocode.columns = [x.replace("Date_of_Incident_(month/day/year)","Date") for x in PK_geocode.columns]
```

```python
PK_geocode['Body_Camera'] = PK_geocode['Body_Camera'].str.replace('no','No')\
.str.replace('Bystander Video','No').str.replace('Surveillance Video','No')
```

```python
PK_geocode.dropna(subset = ['Symptoms_of_mental_illness?'],inplace=True)
PK_geocode.dropna(subset = ['Body_Camera'],inplace=True)
```

```python
PK_geocode['Symptoms_of_mental_illness?'].value_counts()
```

```python
PK_geocode.columns
```

```python
PK_geocode.drop_duplicates(inplace=True)
```

```python
# PK_geocode[PK_geocode['Date'].str.contains("2014")]
PK_geocode['Date'] = PK_geocode['Date'].str.replace('2014','2015')
```

```python
PK_geocode.to_csv('PK_geocode.csv')
```

```python
census = pd.read_csv('censusdata/acs2017_county_data.csv')
```

```python
census = census[census.State != 'Puerto Rico']
```

```python
census.columns
```

```python
census['Unemployment_per_state'] = census.groupby('State')['Unemployment'].transform('mean')
census['Total_Population'] = census.groupby('State')['TotalPop'].transform('sum')

list_of_races = ['Hispanic','White', 'Black', 'Native', 'Asian', 'Pacific']
for race in list_of_races:
    census['{}_percentage'.format(race)] = census.groupby('State')[race].transform('mean')
    census['{}_percentage'.format(race)] = census['{}_percentage'.format(race)].round(2)
```

```python
census = census[['State','Unemployment_per_state',
       'Total_Population', 'Hispanic_percentage', 'White_percentage',
       'Black_percentage', 'Native_percentage', 'Asian_percentage',
       'Pacific_percentage']]
```

```python
census.drop_duplicates(inplace=True)
```
